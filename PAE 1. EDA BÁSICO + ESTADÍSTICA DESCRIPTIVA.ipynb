# üìä <span style="color: #6abe77">Visualizaciones del EDA</span>

## üìä <span style="color:#6abe77">Histogramas de Latitud, Longitud y Altitud de cada circuito</span>

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

numeric_cols = ["lat", "lng", "alt"]

df[numeric_cols].hist(bins=15, figsize=(12, 6), edgecolor='black')
plt.suptitle("Histogramas de datos geogr√°ficos de cada circuito", fontsize=16)
plt.show()
```

### üìä <span style="color:#6abe77">A) Histogramas ‚Äî Interpretaci√≥n</span>

* **üåê Lat (latitud)**
    * La distribuci√≥n **no es normal**.
    * Est√° **sesgada hacia la derecha**, porque la mayor√≠a de latitudes est√°n entre **30¬∞ y 50¬∞** (Europa y Asia).
    * Hay **pocos circuitos** con latitudes negativas (hemisferio sur).

* **‚ÜîÔ∏è Lng (longitud)**
    * La distribuci√≥n es **bimodal o multimodal**:
        * Hay acumulaci√≥n de circuitos tanto en **longitudes negativas** (Am√©rica/Occidente) como **positivas** (Europa/Asia).
    * **No es normal** ni uniforme.

* **‚õ∞Ô∏è Alt (altitud)**
    * Est√° **fuertemente sesgada hacia la derecha**.
    * La mayor√≠a de valores est√°n muy cerca de **0‚Äì200 m**, pero hay pocos circuitos extremadamente altos (hasta m√°s de 2000 m), formando una **cola larga**.

### üì¶ <span style="color:#6abe77">B) Boxplots ‚Äî Interpretaci√≥n</span>

#### ‚ùì **¬øHay valores at√≠picos?**
**S√≠.**

* **‚õ∞Ô∏è Altitud:** Hay varios valores at√≠picos **muy altos**, como el circuito de Ciudad de M√©xico (m√°s de 2000 m), y otros entre 1000 y 1500 m.
* **‚ÜîÔ∏è Longitud:** Hay *outliers* en **ambos extremos** debido a la dispersi√≥n global de los circuitos (Asia, Am√©rica, Medio Oriente vs. Europa).
* **üåê Latitud:** Se observan algunos valores at√≠picos en el **hemisferio sur**, pero menos que en longitud.

#### üìè **¬øC√≥mo se ubica la mediana?**

* **‚õ∞Ô∏è Altitud:** La mediana est√° **muy abajo** dentro de la caja, indicando que la mayor√≠a de circuitos est√°n cerca del nivel del mar y solo unos pocos est√°n mucho m√°s altos.
* **üåê Latitud y Longitud:** La mediana est√° **m√°s centrada** dentro de sus cajas, lo que indica distribuciones m√°s equilibradas aunque con dispersi√≥n global.

### üè≥Ô∏è <span style="color:#6abe77">N√∫mero de circuitos por pa√≠s</span>

### üó∫Ô∏è <span style="color:#6abe77">An√°lisis de dispersi√≥n entre Latitud y Longitud para identificar zonas con mayor concentraci√≥n de eventos</span>

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="lng", y="lat")
plt.title("Mapa Dispersi√≥n de Circuitos (Latitud vs Longitud)", fontsize=16)
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.show()
```

### üó∫Ô∏è <span style="color:#6abe77">¬øSe observa alguna tendencia o patr√≥n general?</span>

**S√≠.** Aunque no hay una correlaci√≥n matem√°tica, el gr√°fico muestra un **patr√≥n geogr√°fico** muy claro:

* üåç La distribuci√≥n forma un **‚Äúmapamundi‚Äù**, ya que la Latitud vs. Longitud ubica los circuitos seg√∫n su posici√≥n real en el planeta.
